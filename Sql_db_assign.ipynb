```python
import sqlite3
import pandas as pd
```

```python
df = pd.read_csv("ODI_Match_info.csv")
df.head()
```

```python
# Connect to (or create) the SQLite database
conn = sqlite3.connect("odi_matches.db")

# Write the data to a table
df.to_sql("match_info", conn, if_exists="replace", index=False)

# Confirm
print("✅ CSV data successfully imported into SQLite database (odi_matches.db)")

# Optional: Verify row count
rows = conn.execute("SELECT COUNT(*) FROM match_info").fetchone()[0]
print(f"Total rows imported: {rows}")

conn.close()
```

```python
from apii import get_api_key
api_key = get_api_key()
```

```python
from langchain_community.utilities.sql_database import SQLDatabase
from langchain_community.agent_toolkits.sql.base import create_sql_agent
from langchain_google_genai import GoogleGenerativeAI
```

```python
DBName= "odi_matches.db"
try:
    db = SQLDatabase.from_uri(f"sqlite:///{DBName}")
except Exception as e:
    print(e)
```

```python
llm = GoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=api_key, temperature=0.2)
```

```python
llm.invoke("hello")
```

```python
agent = create_sql_agent(
    llm = llm , 
    db = db , 
    agent_type="zero-shot-react-description",
    verbose=True
)
```

```python
def chat(question):
    response = agent.invoke(question)
    return response
```

```python
chat("how many times Pakistan won the match as winner")
```
